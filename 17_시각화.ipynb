# scatter, 산점도


```python
# scatter, 산점도

import numpy as np
import matplotlib.pyplot as plt

# ✅ 0부터 9까지 정수 배열 생성 (총 10개)
data = np.arange(10)

# ✅ 선 그래프 (기본 plot)으로 시각화
plt.plot(data)  
plt.title("Line Plot 예제")
plt.show()

```

```python
# ✅ 산점도 그래프 (각 점을 따로 찍음)
plt.scatter(np.arange(10), data)  # x값: 0~9, y값: data
plt.title("Scatter Plot 예제")
plt.show()
```

| 구분          | 설명                             |
| ----------- | ------------------------------ |
| `plot()`    | 선을 이은 꺾은선 그래프 (기본 연결 그래프)      |
| `scatter()` | 점 하나하나만 표시, **분포나 밀집도** 확인에 좋음 |

Seaborn.tips 데이터셋

```python
# seaborn 라이브러리 불러오기 (시각화와 예제 데이터셋 포함됨)
import seaborn as sns

# seaborn의 내장 예제 데이터셋 중 하나인 'tips' 데이터를 불러와 DataFrame으로 저장
tips_df = sns.load_dataset('tips')

# DataFrame의 앞 5줄 미리 보기 (구조와 컬럼을 빠르게 파악할 때 사용)
tips_df.head()
```

```python
# 1번량 그래프 히스토그램 (수치형 데이터 활용)

plt.hist(data=tips_df, x ='total_bill', bins=20)
```

### 📌 형식: fig.add_subplot(nrows, ncols, index)

| 인자      | 의미                                 |
| ------- | ---------------------------------- |
| `nrows` | subplot의 **행(row) 수**              |
| `ncols` | subplot의 **열(column) 수**           |
| `index` | **몇 번째 위치**의 subplot인지 (왼→오, 위→아래) |


```python
import matplotlib.pyplot as plt

# 1. 전체 figure(도화지)를 생성하고 크기를 지정 (가로 12인치, 세로 5인치)
fig = plt.figure(figsize=(12, 5))

# 2. 첫 번째 subplot (1행 2열 중 첫 번째 위치)에 axes 추가
# -> 왼쪽에 'total_bill'을 그릴 공간
axes1 = fig.add_subplot(1, 2, 1)

# 3. 두 번째 subplot (1행 2열 중 두 번째 위치)에 axes 추가
# -> 오른쪽에 'tip'을 그릴 공간
axes2 = fig.add_subplot(1, 2, 2)

# 4. 첫 번째 axes에 'total_bill' 컬럼 기준으로 히스토그램을 그림
# bins=20은 구간을 20개로 나눈다는 뜻
axes1.hist(data=tips_df, x='total_bill', bins=20)
axes1.set_title('Total Bill')  # 그래프 제목 설정

# 5. 두 번째 axes에 'tip' 컬럼 기준으로 히스토그램을 그림
axes2.hist(data=tips_df, x='tip', bins=20)
axes2.set_title('Tip')  # 그래프 제목 설정

# 6. subplot들 사이 간격 자동 조절 (그래프가 겹치지 않도록)
plt.tight_layout()

# 7. 그래프를 화면에 출력
plt.show()
```

# boxplot 차트

## 박스 플롯

상자 그래프라고도 하는 박스 그래프는 이산 변수와 연속 변수 사이의 관계를 나타낸다.  

연속형 변수의 분포를 나타내는 그래프로 최대/최소/중앙값과 사분위수를 표시

1. 제1사분위수(Q1)와 제3사분위수(Q3)를 양 끝으로 하는 상자를 그림.
2. 상자 안에 중앙값을 수평선으로 표시.
3. 상자의 길이, 즉 Q1에서 Q3까지의 범위를 사분범위(IQR, Inter-quartile Range)라고 하며 이 상자 길이의 1.5배 안에 있는 데이터를 기준으로 수염을 표시.
4. 수염 바깥에 있는 데이터는 작은 원으로 표현하며, 이상치(Outliers)로 간주

![Boxplot 이미지](image/boxplot.png)


```python
# boxlot
fig = plt.figure(figsize=(12,5))
axes = fig.add_subplot(1,1,1) #total_bill
axes.boxplot(data=tips_df, x='total_bill')
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. 전체 figure(도화지)를 생성하고 크기를 설정 (가로 12인치, 세로 5인치)
fig = plt.figure(figsize=(12, 5))

# 2. 하나의 subplot(1행 1열 중 첫 번째)에 axes 생성
# 여기서 'axes'는 그래프가 그려질 공간입니다.
axes = fig.add_subplot(1, 1, 1)

# 3. 박스플롯(boxplot) 그리기
axes.boxplot(
    x=[
        # 3-1. 성별이 'Female'인 행의 'total_bill' 값들
        tips_df[tips_df['sex'] == 'Female']['total_bill'],
        # 3-2. 성별이 'Male'인 행의 'total_bill' 값들
        tips_df[tips_df['sex'] == 'Male']['total_bill']
    ],
    # label=['Female', 'Male']  # 각 박스의 레이블 지정
)

# (선택사항) 레이블이 안 보일 경우에는 아래처럼 xtick 설정 추가해도 됩니다:
    # label=['Female', 'Male']  # 각 박스의 레이블 지정 이 코드 대신에
axes.set_xticklabels(['Female', 'Male'])

# 4. 그래프 출력
plt.show()
```

```python

# seaborn의 내장 예제 데이터셋 중 하나인 'tips' 데이터를 불러와 DataFrame으로 저장
tips_df = sns.load_dataset('tips')

# DataFrame의 앞 5줄 미리 보기 (구조와 컬럼을 빠르게 파악할 때 사용)
tips_df.head()
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize=(12, 5))
axes = fig.add_subplot(1, 1, 1)

# 박스플롯: 팁이 1달러 이상인 경우, 시간대별 팁 분포 비교
axes.boxplot(
    x=[
        tips_df[(tips_df['tip'] >= 1) & (tips_df['time'] == 'Lunch')]['tip'],
        tips_df[(tips_df['tip'] >= 1) & (tips_df['time'] == 'Dinner')]['tip']
    ]
)

# x축 라벨 설정
axes.set_xticklabels(['Lunch (tip ≥ 1)', 'Dinner (tip ≥ 1)'])
axes.set_ylabel('Tip Amount ($)')

# 그래프 제목
axes.set_title('Tip Distribution by Time (tip ≥ $1)')

# 레이아웃 자동 정리
axes.yaxis.grid(True)

plt.tight_layout()
plt.show()
```

# scatter, 산점도

```python
colors = {"Female": "red", "Male": "blue"}

# 다변량 표현
# 성별에 따라 - 색상(C 속성)
# size에 따라 - 점 크기(S 속성)

fig = plt.figure()
axes = fig.add_subplot(1, 1, 1)  # total_bill vs tip

# 산점도 그리기
axes.scatter(
    tips_df['total_bill'], tips_df['tip'],
    # s=tips_df['size']**2 * 10, #tablesize > tip금액으로 바꿔보기
    s=tips_df['tip'] * 20,  # ⬅️ tip 금액이 클수록 점 크게
    c=tips_df['sex'].map(colors)
)

# ✅ 격자선 추가
axes.grid(True)

# (선택) 축 라벨과 제목도 추가해보면 더 직관적입니다
axes.set_xlabel('Total Bill ($)')
axes.set_ylabel('Tip ($)')
# axes.set_title('Tip vs Total Bill (Size = Group Size, Color = Sex)')
axes.set_title('Tip vs Total Bill (Size = Tip Amount, Color = Sex)')


plt.tight_layout()
plt.show()
```

```python
# ✅ 예제 1: smoker 여부로 색상 구분하기

colors = {"Yes":"green", "No":"gray"}

fig = plt.figure()
axes = fig.add_subplot(1, 1, 1)

axes.scatter(
    tips_df['total_bill'], tips_df['tip'],
    s=tips_df['size'] * 30,
    c=tips_df['smoker'].map(colors)
)

axes.set_xlabel("Total Bill ($)")
axes.set_ylabel("Tip ($)")
axes.set_title("Smoker vs Tip Amount")
axes.grid(True)

plt.tight_layout()
plt.show
            
```

```python
# ✅ 예제 2: 요일(day)에 따른 색상 구분 + 팁 크기에 따라 점 크기

colors = {"Thur": "purple", "Fri":"green", "Sat":"blue", "Sun":"orange"}

fig = plt.figure()
axes = fig.add_subplot(1,1,1)

axes.scatter(
    tips_df['total_bill'], tips_df['tip'],
    s=tips_df['tip']*25,
    c=tips_df['day'].map(colors)
)

axes.set_xlabel('Total Bill($)')
axes.set_ylabel('Tip ($)')
axes.set_title("Day vs Tip Amount")
axes.grid(True)

plt.tight_layout()
plt.show()
```
