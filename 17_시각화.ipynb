```python
import matplotlib.pyplot as plt

# 👉 전체 그림을 1행 2열로 나눔 (subplot 1번째: 왼쪽 그래프)
plt.subplot(1, 2, 1)
plt.plot(dates, prices)  # 선 그래프 그리기
plt.xticks(rotation=40)  # x축 라벨을 40도 회전해서 보기 쉽게

# 👉 subplot 2번째: 오른쪽 그래프
plt.subplot(1, 2, 2)
plt.bar(dates, prices)   # 막대 그래프 그리기
plt.xticks(rotation=40)  # x축 라벨 회전

# 👉 현재 전체 그래프(2개)를 'my_stock_price_chart.png'라는 이름으로 저장
plt.savefig('my_stock_price_chart.png')
```

# scatter, 산점도


```python
# scatter, 산점도

import numpy as np
import matplotlib.pyplot as plt

# ✅ 0부터 9까지 정수 배열 생성 (총 10개)
data = np.arange(10)

# ✅ 선 그래프 (기본 plot)으로 시각화
plt.plot(data)  
plt.title("Line Plot 예제")
plt.show()

```

```python
# ✅ 산점도 그래프 (각 점을 따로 찍음)
plt.scatter(np.arange(10), data)  # x값: 0~9, y값: data
plt.title("Scatter Plot 예제")
plt.show()
```

| 구분          | 설명                             |
| ----------- | ------------------------------ |
| `plot()`    | 선을 이은 꺾은선 그래프 (기본 연결 그래프)      |
| `scatter()` | 점 하나하나만 표시, **분포나 밀집도** 확인에 좋음 |
