## 조건식의 규칙

- 조건식으로 사용할 행렬의 값이 문자여도 결과만 True/False 식으로 나오면 필터가 된다.
- 필터로 사용할 행렬의 갯수 == 원본 데이터의 갯수 : "짝이 맞아야 한다" Indexerr
- 행렬[조건식, ]
- 행렬[:, 조건식]
- 괄호( ) 로 우선순위를 지정할 수 있다. 
- 복합조건 : 논리함수 and, or > &, | 로 표현한다.
- 부정연산자 : != , ~
