# í”¼í´, ì§ë ¬í™” / ì—­ì§ë ¬í™”

ì§ë ¬í™”(Serialization)ëŠ” íŒŒì´ì¬ì˜ ê°ì²´ë‚˜ ë°ì´í„° êµ¬ì¡°ë¥¼ íŒŒì¼ì´ë‚˜ ë„¤íŠ¸ì›Œí¬ë¡œ ì „ì†¡í•  ìˆ˜ ìˆëŠ” í˜•ì‹ìœ¼ë¡œ ë³€í™˜í•˜ëŠ” ê³¼ì •ì´ë‹¤.
ë°˜ëŒ€ë¡œ ì—­ì§ë ¬í™”(Deserialization)ëŠ” íŒŒì¼ì´ë‚˜ ë„¤íŠ¸ì›Œí¬ì—ì„œ ë°›ì€ ë°ì´í„°ë¥¼ ë‹¤ì‹œ íŒŒì´ì¬ ê°ì²´ë¡œ ë³€í™˜í•˜ëŠ” ê³¼ì •ì„ ë§í•œë‹¤.

Pythonì—ì„œëŠ” ì£¼ë¡œ `pickle` ëª¨ë“ˆì„ ì‚¬ìš©í•˜ì—¬ ì§ë ¬í™”ì™€ ì—­ì§ë ¬í™”ë¥¼ ìˆ˜í–‰í•œë‹¤.

ì§ë ¬í™”ì™€ ì—­ì§ë ¬í™”ì˜ ì£¼ìš” ì‚¬ìš© ì‚¬ë¡€ëŠ” ë‹¤ìŒê³¼ ê°™ë‹¤:

- ê°ì²´ì˜ ì˜ì†ì„±(persistence) ìœ ì§€
- ë„¤íŠ¸ì›Œí¬ë¥¼ í†µí•œ ë°ì´í„° ì „ì†¡
- ë³µì¡í•œ ë°ì´í„° êµ¬ì¡°ì˜ ì €ì¥ê³¼ ë³µì›

### ë‹¤ìŒì€ pickle ëª¨ë“ˆì„ ì‚¬ìš©í•œ ì§ë ¬í™”/ì—­ì§ë ¬í™”ì˜ ê¸°ë³¸ ì˜ˆì œ

````python
import pickle

# ì§ë ¬í™”í•  ë°ì´í„° ë”•ì…”ë„ˆë¦¬
data = {
    'name': 'í™ê¸¸ë™',
    'age': 30,
    'scores': [90, 85, 88]
}

# ì§ë ¬í™”í•˜ì—¬ íŒŒì¼ì— ì €ì¥
with open('abc\íŒŒì¼ì´ë¦„.pickle', 'wb') as f:
    pickle.dump(data, f)

# íŒŒì¼ì—ì„œ ì—­ì§ë ¬í™”í•˜ì—¬ ì½ê¸°
with open('abc\íŒŒì¼ì´ë¦„.pickle', 'rb') as f:
    loaded_data = pickle.load(f)

print(loaded_data)  # ì›ë³¸ ë°ì´í„°ê°€ ê·¸ëŒ€ë¡œ ë³µì›ë¨
````

ìœ„ ì˜ˆì œì—ì„œ dictionary ê°ì²´ë¥¼ pickleì„ ì‚¬ìš©í•˜ì—¬ íŒŒì¼ì— ì €ì¥í•˜ê³ , ë‹¤ì‹œ ì½ì–´ì˜¤ëŠ” ê³¼ì •ì„ ë³´ì—¬ì¤€ë‹¤. ì§ë ¬í™”ëœ ë°ì´í„°ëŠ” ë°”ì´ë„ˆë¦¬ í˜•íƒœë¡œ ì €ì¥ë˜ë¯€ë¡œ, íŒŒì¼ì„ ì—´ ë•Œ 'wb'(ì“°ê¸°-ë°”ì´ë„ˆë¦¬) ë˜ëŠ” 'rb'(ì½ê¸°-ë°”ì´ë„ˆë¦¬) ëª¨ë“œë¥¼ ì‚¬ìš©í•´ì•¼ í•œë‹¤.

````python
import json

# ì§ë ¬í™”í•˜ì—¬ íŒŒì¼ì— ì €ì¥
with open('abc\íŒŒì¼ì´ë¦„.json', 'w') as f:
    json.dump(data, f, ensure_ascii=False)
````

````python
with open('abc\íŒŒì¼ì´ë¦„.json', 'r') as f:
    loaded_txt = json.load(f)

print(loaded_txt)
````
