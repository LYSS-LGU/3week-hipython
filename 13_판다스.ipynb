````python
import pandas as pd

# ê¸°ì¡´ DataFrame
data = {
    'name': ['ê¹€ì² ìˆ˜', 'ì´ì˜í¬', 'ë°•ë¯¼ìˆ˜'],
    'age': [25, 28, 22],
    'score': [85, 92, 78]
}
df = pd.DataFrame(data)

# ìƒˆ í–‰ ë§Œë“¤ê¸°
new_row = pd.DataFrame([['ê¹€ì§€ìˆ˜', 23, 95]], columns=['name', 'age', 'score'])

# concatìœ¼ë¡œ í–‰ ì¶”ê°€
df = pd.concat([df, new_row], ignore_index=True)

print(df)
````
